###  Polynomial Regresion to Multiple-linear Regression

---

    When data do not have a linear relationship between the independent variables x and response variable y (or the outcome), complicated models might fit in the data. 
    
    The next possible model is the Polynomial models. Actually we can convert polynomial relation into multi-linear relation. Thus, we are still working on linear regression. 
    
    Suppose we use the following polynomial equation to match the data:

$$ y = b + w_{1}*x + w_{2}x^{2} + .... + w_{d}x^{d} $$

    Where d is the degree of the polynomial.

    What we can do is data transformation.   
    transformation:  
$$   x  -> x_{1} $$  
$$   x^{2}  -> x_{2} $$  
$$ .......... $$  
$$   x^{d}  -> x_{d} $$  

    Now the problem becomes:
$$ y = b + w_{1}*x_{1} + w_{2}x_{2} + .... + w_{d}x_{d} $$
  
  
    This is again another multiple-linear regression problem.

## import some modules

```python
# Import python modules 
import math
import numpy as np
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

## 0. load data into memory and visualize the data

```python
# Read in train and test synthetic data from local storage
train_x = np.loadtxt('data_train.txt', delimiter = ',', dtype=float)
train_y = np.loadtxt('label_train.txt', delimiter = ',', dtype=float)

test_x =  np.loadtxt('data_test.txt', delimiter = ',', dtype=float)
test_y =  np.loadtxt('label_test.txt', delimiter = ',', dtype=float)

# (1) get data and labels
train_x = train_x.reshape(-1,1) # data changed to column vector
test_x = test_x.reshape(-1,1) # data changed to column vector

# show the data in plotting
plt.scatter(train_x, train_y, color='blue', label = "Training data")
plt.scatter(test_x, test_y,  color='red', label = "Testing data")
plt.legend()
plt.show()
```

## 1. Data transformation, then Linear Regression (10 points)  
   
   * For a given degree d = 2
       - print out the error MSE   
       - plot the y_prediction vs y_test
   
   * For given degree d = 10 and d = 15 
       - do the same
     

```python
# (a) transform data to higher order using the preprocessing module: 
# PolynomialFeatures(degree=d), d is user-defined, for example d = 2, 10, 15


d2=PolynomialFeatures(degree=2)


d2train_x=train_x.copy()

d2test_x=test_x.copy()

d2train_x=d2.fit_transform(d2train_x) 
d2test_x=d2.fit_transform(d2test_x)
# (b) Apply linear models: LinearRegression()
# Train it with the transformed data

LRd2 = LinearRegression().fit(d2train_x, train_y)

# (c) Test the model with transformed testing data 

LRd2pred=LRd2.predict(d2test_x)



# (d) Compute the error MSE


diff=np.subtract(LRd2pred, test_y)
squared=np.square(diff)
d2MSE=squared.mean()
print("The MSE for a degree of 2 is "+str(d2MSE))



# (e) Plotting the prediction y_hat and  y_test values

#y_hat == the predicted y values??

plt.scatter(test_x, test_y, color='blue', label= 'Test Data')
plt.scatter(test_x, LRd2pred, color='red', label='Predictions')
plt.legend()
plt.show()
```

```python
# (a) transform data to higher order using the preprocessing module: 
# PolynomialFeatures(degree=d), d is user-defined, for example d = 2, 10, 15
d10=PolynomialFeatures(degree=10)

d10train_x=train_x.copy()

d10test_x=test_x.copy()
d10train_x=d10.fit_transform(d10train_x) 
d10test_x=d10.fit_transform(d10test_x)
# (b) Apply linear models: LinearRegression()
# Train it with the transformed data

LRd10 = LinearRegression().fit(d10train_x, train_y)

# (c) Test the model with transformed testing data 

LRd10pred=LRd10.predict(d10test_x)



# (d) Compute the error MSE


diff=np.subtract(LRd10pred, test_y)
squared=np.square(diff)
d10MSE=squared.mean()
print("The MSE for a degree of 10 is "+str(d10MSE))



# (e) Plotting the prediction y_hat and  y_test values

#y_hat == the predicted y values??

plt.scatter(test_x, test_y, color='blue', label= 'Test Data')
plt.scatter(test_x, LRd10pred, color='red', label='Predictions')
plt.legend()
plt.show()
```

```python
# (a) transform data to higher order using the preprocessing module: 
# PolynomialFeatures(degree=d), d is user-defined, for example d = 2, 10, 15
d15=PolynomialFeatures(degree=15)

d15train_x=train_x.copy()

d15test_x=test_x.copy()
d15train_x=d15.fit_transform(d15train_x) 
d15test_x=d15.fit_transform(d15test_x)
# (b) Apply linear models: LinearRegression()
# Train it with the transformed data

LRd15 = LinearRegression().fit(d15train_x, train_y)

# (c) Test the model with transformed testing data 

LRd15pred=LRd15.predict(d15test_x)



# (d) Compute the error MSE


diff=np.subtract(LRd15pred, test_y)
squared=np.square(diff)
d15MSE=squared.mean()
print("The MSE for a degree of 15 is "+str(d15MSE))



# (e) Plotting the prediction y_hat and  y_test values

#y_hat == the predicted y values??

plt.scatter(test_x, test_y, color='blue', label= 'Test Data')
plt.scatter(test_x, LRd15pred, color='red', label='Predictions')
plt.legend()
plt.show()
```

## 2. Learning curve:  MSE  ~ d (5 points)

   - plot the learning curve:  MSE ~ d (or Error ~ degree)

```python
# plot the learning curve
MSElist=[]
dlist=[]
for i in range(2,20):
    dlist.append(i)
    
    #creating the transformed x variable arrays
    thisd=PolynomialFeatures(degree=i)
    thistrain_x=train_x.copy()
    thistest_x=test_x.copy()
    thistrain_x=thisd.fit_transform(thistrain_x)
    thistest_x=thisd.fit_transform(thistest_x)
    
    #fitting the model to the current degree
    thisLR=LinearRegression().fit(thistrain_x, train_y)
    
    #making a prediction off of the current degree
    thisLRpred=thisLR.predict(thistest_x)
    
    #finding the MSE of the current degree
    diff=np.subtract(thisLRpred, test_y)
    squared=np.square(diff)
    thisMSE=squared.mean()
    MSElist.append(thisMSE)

#plotting
plt.plot(dlist, MSElist)
plt.show()

MSEsmall=min(MSElist)
smallIndex=MSElist.index(MSEsmall)
print("Smallest MSE: "+str(MSEsmall)+" Degree with Smallest MSE: "+str(smallIndex+1))











```
